```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory="myvectors",
    collection_name="sample",
)
```

```python
d1 = Document(
    page_content="Winning the art competition made her incredibly happy.",
    metadata={"context": "part of story"},
)

d2 = Document(
    page_content="The café on the corner serves really good coffee.",
    metadata={"context": "feedback of cafe"},
)

d3 = Document(
    page_content="The train arrives at the station every hour",
    metadata={"context": "saying of  train master"},
)

d4 = Document(
    page_content="He felt frustrated after losing his wallet.",
    metadata={"context": "every time"},
)
```

```python
d = [d1, d2, d3, d4]
```

```python
vector_store.add_documents(d)
```

```python
vector_store.get(include=['metadatas','embeddings','documents'])
```

```python
vector_store.similarity_search(query='loosing a competition make person strong.',k=1)
```

```python
update_doc = Document(
    page_content='This is the first time she has won the art competition.',
    metadata={'context':'partof story'}
)
```

```python
vector_store.update_document(document_id='d11541ee-d146-4508-b729-eb6d1559e785',document=update_doc)
```
